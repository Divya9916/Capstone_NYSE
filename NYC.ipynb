```python
# 1 ADJUSTED STOCK PRICES
```

```python
# Top 10 Companies from Fundamentals
top10 = fund.groupby('Ticker Symbol')['Total Revenue'].mean().sort_values(ascending=False).head(10).index
```

```python
# Filter Price Data
top10_prices = Price_adj[Price_adj['symbol'].isin(top10)]
```

```python
# Plot Adjusted Closing Prices (Top 10 Companies)
plt.figure(figsize=(14,7))

sns.lineplot(
    data=top10_prices,
    x='date',
    y='close',
    hue='symbol',
    linewidth=2
)

plt.title('Adjusted Closing Price Trend Across Companies')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Adjusted Closing Price', fontsize=12)

plt.grid(True)
plt.show()
```
